```python
import numpy as np
import pandas as pd  
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.DataFrame()
df = pd.read_csv("C:/Users/aath/Dropbox/MAEN/Thankful/Data/FLS_TypeSale.csv")
plt.figure(figsize=(10, 5))



plt.subplot(2, 2, 1)
plt.scatter(x=df['Year'], y=df['HOA'], color='purple')
plt.ylabel('Trustee No Listing')
plt.xlabel('year')
plt.title('trustee HOA sale over years')

plt.subplot(2, 2, 2)
plt.scatter(x=df['Year'], y=df['TaxT'], color='purple')
plt.ylabel('Trustee No Listing')
plt.xlabel('year')
plt.title('trustee TaxT sale over years')

plt.subplot(2, 2, 3)
plt.scatter(x=df['Year'], y=df['HE'], color='purple')
plt.ylabel('Trustee No Listing')
plt.xlabel('year')
plt.title('trustee HE sale over years')


plt.subplot(2, 2, 4)
plt.scatter(x=df['Year'], y=df['FHA'], color='purple')
plt.ylabel('Trustee No Listing')
plt.xlabel('year')


plt.title('trustee FHA sale over years')

plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1)

plt.show()
```

```python
import numpy as np
import pandas as pd  
import matplotlib.pyplot as plt
%matplotlib inline

# Import foreclosure data from C:

df = pd.DataFrame()
df = pd.read_csv("C:/Users/aath/Dropbox/MAEN/Thankful/Data/FLS_TypeSale.csv")

# Prepare to plot 4 vectors of data over time by 4 subplots

plt.figure(figsize=(10, 5))

plt.subplot(2, 2, 1)
plt.scatter(x=df['Year'], y=df['HOA'], color='purple')
plt.ylabel('Trustee No Listing')
plt.xlabel('year')
plt.title('trustee HOA sale over years')

plt.subplot(2, 2, 2)
plt.scatter(x=df['Year'], y=df['TaxT'], color='purple')
plt.ylabel('Trustee No Listing')
plt.xlabel('year')
plt.title('trustee TaxT sale over years')

plt.subplot(2, 2, 3)
plt.scatter(x=df['Year'], y=df['HE'], color='purple')
plt.ylabel('Trustee No Listing')
plt.xlabel('year')
plt.title('trustee HE sale over years')

plt.subplot(2, 2, 4)
plt.scatter(x=df['Year'], y=df['FHA'], color='purple')
plt.ylabel('Trustee No Listing')
plt.xlabel('year')

plt.title('trustee FHA sale over years')

plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1)

plt.show()




```
