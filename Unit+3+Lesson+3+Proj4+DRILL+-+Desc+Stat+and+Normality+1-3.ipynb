```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#**************************************************************************************************************************
#                                            gamma Distribution
#**************************************************************************************************************************

# Generate the distribution

gamma = np.random.gamma(5,1, 100)

# Plot a histogram
plt.hist(gamma)

# Add a vertical line at the mean.
plt.axvline(gamma.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(gamma.mean() + gamma.std(), color='b', linestyle='dashed', linewidth=2)

#Add a vertical line at one standard deviation below the mean.
plt.axvline(gamma.mean() - gamma.std(), color='b', linestyle='dashed', linewidth=2)

plt.title('gamma Distribution')

# Print the histogram
plt.show()

gammaStd = np.std(gamma, ddof=1)
gammaMean = np.mean(gamma)
gammaMedian = np.median(gamma)
print("gamma St Dev = {},".format(gammaStd), ("gamma Mean = {},".format(gammaMean)),
     ("gamma Median = {},".format(gammaMedian)))
   

#**************************************************************************************************************************
#                                            binomial Distribution
#**************************************************************************************************************************

# Generate the distribution

binomial = np.random.binomial(2,0.5, 100)

# Plot a histogram
plt.hist(binomial)

# Add a vertical line at the mean.
plt.axvline(binomial.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(binomial.mean() + binomial.std(), color='b', linestyle='dashed', linewidth=2)

#Add a vertical line at one standard deviation below the mean.
plt.axvline(binomial.mean() - binomial.std(), color='b', linestyle='dashed', linewidth=2)

plt.title('binomial Distribution')

# Print the histogram
plt.show()

binomialStd = np.std(binomial, ddof=1)
binomialMean = np.mean(binomial)
binomialMedian = np.median(binomial)
print("binomial St Dev = {},".format(binomialStd), ("binomial Mean = {},".format(binomialMean)),
     ("binomial Median = {},".format(binomialMedian)))

#**************************************************************************************************************************
#                                            lognormal Distribution
#**************************************************************************************************************************

# Generate the distribution

lognormal = np.random.lognormal(10,1, 100)

# Plot a histogram
plt.hist(lognormal)

# Add a vertical line at the mean.
plt.axvline(lognormal.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(lognormal.mean() + lognormal.std(), color='b', linestyle='dashed', linewidth=2)

#Add a vertical line at one standard deviation below the mean.
plt.axvline(lognormal.mean() - lognormal.std(), color='b', linestyle='dashed', linewidth=2)

plt.title('lognormal Distribution')

# Print the histogram
plt.show()

lognormalStd = np.std(lognormal, ddof=1)
lognormalMean = np.mean(lognormal)
lognormalMedian = np.median(lognormal)
print("lognormal St Dev = {},".format(lognormalStd), ("lognormal Mean = {},".format(lognormalMean)),
     ("lognormal Median = {},".format(lognormalMedian)))

#**************************************************************************************************************************
#                                            poisson Distribution
#**************************************************************************************************************************

# Generate the distribution

poisson = np.random.poisson(1, 100)

# Plot a histogram
plt.hist(poisson)

# Add a vertical line at the mean.
plt.axvline(poisson.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(poisson.mean() + poisson.std(), color='b', linestyle='dashed', linewidth=2)

#Add a vertical line at one standard deviation below the mean.
plt.axvline(poisson.mean() - poisson.std(), color='b', linestyle='dashed', linewidth=2)

plt.title('poisson Distribution')

# Print the histogram
plt.show()

poissonStd = np.std(poisson, ddof=1)
poissonMean = np.mean(poisson)
poissonMedian = np.median(poisson)
print("poisson St Dev = {},".format(poissonStd), ("poisson Mean = {},".format(poissonMean)),
     ("poisson Median = {},".format(poissonMedian)))


#**************************************************************************************************************************
#                                            triangular Distribution
#**************************************************************************************************************************

# Generate the distribution

triangular = np.random.triangular(-3, 0, 3, 100)

# Plot a histogram
plt.hist(triangular)

# Add a vertical line at the mean.
plt.axvline(triangular.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(triangular.mean() + triangular.std(), color='b', linestyle='dashed', linewidth=2)

#Add a vertical line at one standard deviation below the mean.
plt.axvline(triangular.mean() - triangular.std(), color='b', linestyle='dashed', linewidth=2)

plt.title('triangular Distribution')

# Print the histogram
plt.show()

triangularStd = np.std(triangular, ddof=1)
triangularMean = np.mean(triangular)
triangularMedian = np.median(triangular)
print("triangular St Dev = {},".format(triangularStd), ("triangular Mean = {},".format(triangularMean)),
     ("triangular Median = {},".format(triangularMedian)))

#**************************************************************************************************************************
#                                            normal Distribution
#**************************************************************************************************************************

# Generate the distribution

normal = np.random.normal(10,1, 100)

# Plot a histogram
plt.hist(normal)

# Add a vertical line at the mean.
plt.axvline(normal.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(normal.mean() + normal.std(), color='b', linestyle='dashed', linewidth=2)

#Add a vertical line at one standard deviation below the mean.
plt.axvline(normal.mean() - normal.std(), color='b', linestyle='dashed', linewidth=2)

plt.title('normal Distribution')

# Print the histogram
plt.show()

normalStd = np.std(normal, ddof=1)
normalMean = np.mean(normal)
normalMedian = np.median(normal)
print("normal St Dev = {},".format(normalStd), ("normal Mean = {},".format(normalMean)),
     ("normal Median = {},".format(normalMedian)))
```
